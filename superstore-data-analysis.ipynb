# Task 2: Data Cleaning

In this section, we clean the dataset by handling missing values and correcting data types to ensure accurate analysis.

## Checking Missing Values

We identify missing values in each column.

```python
df.isnull().sum()
```

All columns have 0 missing values, indicating that the dataset is complete and does not require any missing value treatment.

## Converting Data Types

We convert date columns to proper datetime format.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.duplicated().sum()
```

# Task 3: Feature Engineering

In this section, we create new features from existing columns to enhance analysis and extract deeper insights.

## Extracting Month and Year

We extract the month and year from the Order Date to enable time-based analysis.

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Month Name'] = df['Order Date'].dt.month_name()
```

## Creating New Feature: Sales per Unit

We create a new column to calculate sales per unit by dividing Sales by Quantity.

```python
df['Sales per Unit'] = df['Sales'] / df['Quantity']
```

```python
df[['Sales', 'Quantity', 'Sales per Unit']].head()
```

## Verifying New Feature

We verify whether the new column is logically correct.

# Task 4: GroupBy & Aggregation

In this section, we use groupby operations to aggregate data and extract meaningful insights such as total sales, profit distribution, and customer performance.

## Total Sales by Category

We calculate total sales for each product category.

```python
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sales_by_category
```

## Total Profit by Region

We calculate total profit across different regions.

```python
profit_by_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
profit_by_region
```

## Top 5 Customers by Sales

We identify the top 5 customers based on total sales.

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(5)
top_customers
```

## Monthly Sales Trend

We analyze how sales change over time on a monthly basis.

```python
monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = monthly_sales['Order Year'].astype(str) + '-' + monthly_sales['Order Month'].astype(str)
monthly_sales
```

# Task 5: Pivot Tables

In this section, we use pivot tables to summarize the dataset in a structured format for better comparison across categories, regions, and time periods.

## Sales by Category and Region

We create a pivot table to analyze how sales vary across categories and regions.

```python
pivot_category_region = pd.pivot_table(
    df,
    values='Sales',
    index='Category',
    columns='Region',
    aggfunc='sum'
)

pivot_category_region
```

## Sales Trends by Segment Over Time

We analyze how sales change over time across different customer segments.

```python
pivot_segment_time = pd.pivot_table(
    df,
    values='Sales',
    index=['Order Year', 'Order Month'],
    columns='Segment',
    aggfunc='sum'
)

pivot_segment_time
```

## Profit by Sub-Category

We analyze which sub-categories generate the most and least profit.

```python
pivot_profit_subcategory = pd.pivot_table(
    df,
    values='Profit',
    index='Sub-Category',
    aggfunc='sum'
).sort_values(by='Profit', ascending=False)

pivot_profit_subcategory
```

## Splitting the Dataset

We split the dataset into two DataFrames:
- One containing customer and order information
- One containing product and sales information

```python
df_customer = df[['Order ID', 'Customer ID', 'Customer Name', 'Segment', 'Region', 'Order Date']]

df_product = df[['Order ID', 'Product ID', 'Category', 'Sub-Category', 'Sales', 'Quantity', 'Profit']]
```

## Merging the Dataset

We merge the two DataFrames back using Order ID to ensure data consistency.

```python
merged_df = pd.merge(df_customer, df_product, on='Order ID', how='inner')
merged_df.head()
```

## Verifying the Merged Dataset

We check whether the merged dataset maintains structure and integrity.

```python
merged_df.shape
merged_df.info()
```

# Task 8: Insights & Conclusion

In this section, we summarize key findings from the analysis and highlight important business insights derived from the Superstore dataset.

## Most Profitable Category

We identify the category that generates the highest total profit.

```python
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

The most profitable category is **Technology**, indicating high-value products contribute significantly to overall profit.

## Least Performing Region

We identify the region with the lowest total profit.

```python
df.groupby('Region')['Profit'].sum().sort_values()
```

The least performing region is **Central**, showing potential inefficiencies or higher discounts affecting profitability.

## Relationship Between Sales and Profit

We analyze whether higher sales always lead to higher profit.

```python
df[['Sales', 'Profit']].corr()
```

Sales and profit show a moderate positive correlation, but high sales do not always guarantee high profit due to discounts and operational costs.

## Trends Over Time

We observe how sales evolve across months and years.

```python
df.groupby(['Order Year', 'Order Month'])['Sales'].sum()
```

Sales show seasonal variation, with certain months consistently performing better, indicating seasonal demand patterns.

## Final Conclusion

The analysis reveals that Technology is the most profitable category, while the Central region underperforms in profit. Sales and profit are not perfectly aligned due to discounts and operational factors. Additionally, sales trends indicate seasonal behavior, which can be leveraged for better business planning.
